```python
import pandas as pd
import seaborn as sns 
import numpy as np
import glob
import matplotlib.pyplot as plt
from scipy.stats import sem
%matplotlib inline
%config Completer.use_jedi = False
```

```python
#Import FPKM data for cancer types

BRCA_file = '/Users/jake/OneDrive - University of Glasgow/Project/BRCA/GeneExpression_MAF/BRCA_FPKM_protein_all_matched.csv'
KIRC_file = '/Users/jake/OneDrive - University of Glasgow/Project/KIRC/GeneExpression_MAF/KIRC_FPKM_protein_all_matched.csv'
LUAD_file = '/Users/jake/OneDrive - University of Glasgow/Project/LUAD/GeneExpression_MAF/LUAD_FPKM_protein_all_matched.csv'
THCA_file = '/Users/jake/OneDrive - University of Glasgow/Project/THCA/GeneExpression_MAF/THCA_FPKM_protein_all_matched.csv'


BRCA = pd.read_csv(BRCA_file,header=[0,1],sep='\t',index_col=[0])
KIRC = pd.read_csv(KIRC_file,header=[0,1],sep='\t',index_col=[0])
LUAD = pd.read_csv(LUAD_file,header=[0,1],sep='\t',index_col=[0])
THCA = pd.read_csv(THCA_file,header=[0,1],sep='\t',index_col=[0])
```

```python
#Add new Cancer_type column for dispalying the cancer type for each


BRCA['Cancer_type','Cancer_type'] = 'BRCA'
KIRC['Cancer_type','Cancer_type'] = 'KIRC'
LUAD['Cancer_type','Cancer_type'] = 'LUAD'
THCA['Cancer_type','Cancer_type'] = 'THCA'
```

```python
#Create for loop that iterates over each cancer dataframe and adds new cancer_status col

for df in BRCA,KIRC,LUAD,THCA:
    df['Cancer_tissue_status','Cancer_tissue_status'] = df['Cancer_type','Cancer_type'] + '_' + df['Cancer','Cancer'].astype(str)
```

```python
#Concat all cancers together

all_cancers = pd.concat([BRCA,KIRC,LUAD,THCA])
all_cancers
```

```python
#Drop the cancer type col and cancer col incase of data leakage

all_cancers.drop(['Cancer_type','Cancer'],axis=1,level=0,inplace=True)
```

```python
#Check for missing values

all_cancers.isnull().sum().sort_values(ascending=False)
```

```python
#Drop missing values for now ... although maybe to look at working with

all_cancers.dropna(axis=1,inplace=True)
```
