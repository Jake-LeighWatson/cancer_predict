```python
import pandas as pd
import seaborn as sns 
import numpy as np
import glob
import matplotlib.pyplot as plt
%matplotlib inline
%config Completer.use_jedi = False
```

```python
#Import eQTL files for required cancer type

#Add path to eQTL files
path = r'/Users/jake/OneDrive - University of Glasgow/Project/BRCA/eQTL'

all_files = glob.glob(path + "/*.txt")

#For loop to read in eQTL files        
for filename in all_files:
    
    if 'cis' in filename:
        cis_eQTL = pd.read_csv(filename,sep='\t')
    elif 'trans' in filename:
        trans_eQTL = pd.read_csv(filename, sep='\t')
```

```python
cis_eQTL.head()
```

```python
#Create new SNP columsn by splitting the SNP column and also remove numbers from gene names

cis_eQTL['Gene_symbol'] = cis_eQTL['gene'].apply(lambda gene: gene.split('|')[0])
cis_eQTL['Entrez_ID'] = cis_eQTL['gene'].apply(lambda gene: gene.split('|')[1])
cis_eQTL['cis_SNP_ID'] = cis_eQTL['SNP'].apply(lambda x: x.split(':')[0])
cis_eQTL['SNP_Position_Chr (hg19)'] = cis_eQTL['SNP'].apply(lambda x: x.split(':')[1])
cis_eQTL['SNP_Allele'] = cis_eQTL['SNP'].apply(lambda x: x.split(':')[2] + '/' + x.split(':')[3])

#Drop old SNP and gene col
cis_eQTL.drop(['SNP','gene'],axis =1 ,inplace=True)
```

```python
cis_eQTL
```

```python
#Now repeat process for the trans_eQTLS

trans_eQTL['Gene_symbol'] = trans_eQTL['gene'].apply(lambda gene: gene.split('|')[0])
trans_eQTL['Entrez_ID'] = trans_eQTL['gene'].apply(lambda gene: gene.split('|')[1])
trans_eQTL['trans_SNP_ID'] = trans_eQTL['SNP'].apply(lambda x: x.split(':')[0])
trans_eQTL['SNP_Position_Chr (hg19)'] = trans_eQTL['SNP'].apply(lambda x: x.split(':')[1])
trans_eQTL['SNP_Allele'] = trans_eQTL['SNP'].apply(lambda x: x.split(':')[2] + '/' + x.split(':')[3])

#Drop old SNP and gene col
trans_eQTL.drop(['SNP','gene'],axis =1 ,inplace=True)
```

```python
#Write the cis and trans tables to csv

cis_eQTL.to_csv(path.split('/')[5]+'_cis_eQTL.csv')
trans_eQTL.to_csv(path.split('/')[5]+'_trans_eQTL.csv')
```
