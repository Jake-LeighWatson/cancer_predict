```python
import pandas as pd
import seaborn as sns 
import numpy as np
import glob
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Import FPKM and MAF files for required cancer type

#Add path to MAF and FPKM data colleted from TCGA
path = r'/Users/jake/OneDrive - University of Glasgow/Project/BRCA/R_data'

#all csv files
all_files = glob.glob(path + "/*.csv")

#for loop through to read in tumour FPKM, normal FPKM and MAF files
for filename in all_files:
    
    if 'tumour' in filename:
        tumour_FPKM = pd.read_csv(filename,index_col=0)
    elif 'normal' in filename:
        normal_FPKM = pd.read_csv(filename,index_col=0)
    elif 'MAF' in filename:
        MAF = pd.read_csv(filename,low_memory=False,index_col=0)
```

```python
#Set index as 'Gene_symbol' for both tumour and normal FPKM

tumour_FPKM.set_index('Gene_symbol',inplace=True)
normal_FPKM.set_index('Gene_symbol',inplace=True)
```

```python
#Transpose both tumour and normal FPKM datatframes

tumour_FPKM = tumour_FPKM.transpose()
normal_FPKM = normal_FPKM.transpose()
```

```python
#Add a 'Cancer' column to each dataframe 
#with a value of 1 (has cancer) and a value of 0 (does'nt have cancer) in tumour and normal DF, respectively.

tumour_FPKM['Cancer'] = 1
normal_FPKM['Cancer'] = 0
```
