```python
#Filenames

#MAF protein coding files
MAF_VEP_protein_tumour_file = '/Users/jake/OneDrive - University of Glasgow/Project/COAD/FPKM_MAF_notebook/COAD_protein_coding_tumour.csv'
MAF_VEP_protein_normal_file = '/Users/jake/OneDrive - University of Glasgow/Project/COAD/FPKM_MAF_notebook/COAD_protein_coding_normal.csv'

#Gene expression FPKM files
FPKM_tumour_file = '/Users/jake/OneDrive - University of Glasgow/Project/COAD/FPKM_MAF_notebook/COAD_FPKM_tumour.csv'
FPKM_normal_file = '/Users/jake/OneDrive - University of Glasgow/Project/COAD/FPKM_MAF_notebook/COAD_FPKM_normal.csv'

#eQTL files
cis_eQTL_file = '/Users/jake/OneDrive - University of Glasgow/Project/COAD/eQTL/COAD_tumor.cis_eQTL.txt'
trans_eQTL_file = '/Users/jake/OneDrive - University of Glasgow/Project/COAD/eQTL/COAD_tumor.trans_eQTL.txt'
```

```python
#Load cis_eQTL and trans_eQTL files

cis_eQTL = pd.read_csv(cis_eQTL_file,sep='\t')
trans_eQTL = pd.read_csv(trans_eQTL_file,sep='\t')
```

```python

cis_eQTL.head()
```

```python
#Create new SNP columsn by splitting the SNP column and also remove numbers from gene names

cis_eQTL['Gene_symbol'] = cis_eQTL['gene'].apply(lambda gene: gene.split('|')[0])

cis_eQTL['cis_SNP_ID'] = cis_eQTL['SNP'].apply(lambda x: x.split(':')[0])
cis_eQTL['SNP_Position_Chr (hg19)'] = cis_eQTL['SNP'].apply(lambda x: x.split(':')[1])
cis_eQTL['SNP_Allele'] = cis_eQTL['SNP'].apply(lambda x: x.split(':')[2] + '/' + x.split(':')[3])

#Drop old SNP and gene col
cis_eQTL.drop(['SNP','gene'],axis =1 ,inplace=True)
```

```python
#Now repeat process for the trans_eQTLS

trans_eQTL['Gene_symbol'] = trans_eQTL['gene'].apply(lambda gene: gene.split('|')[0])

trans_eQTL['trans_SNP_ID'] = trans_eQTL['SNP'].apply(lambda x: x.split(':')[0])
trans_eQTL['SNP_Position_Chr (hg19)'] = trans_eQTL['SNP'].apply(lambda x: x.split(':')[1])
trans_eQTL['SNP_Allele'] = trans_eQTL['SNP'].apply(lambda x: x.split(':')[2] + '/' + x.split(':')[3])

#Drop old SNP and gene col
trans_eQTL.drop(['SNP','gene'],axis =1 ,inplace=True)
```
